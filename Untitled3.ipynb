```python
import tensorflow as tf
from tensorflow.keras import models, layers, losses
from tensorflow.keras.applications import VGG19
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras import backend as K
import numpy as np
import matplotlib.pyplot as plt

# Load the pre-trained VGG19 model
base_model = VGG19(weights='imagenet', include_top=False)

# Create a model that extracts style and content features
style_layers = ['block1_conv1', 'block2_conv1', 'block3_conv1', 'block4_conv1', 'block5_conv1']
content_layers = ['block5_conv2']

def get_model():
    base_model.trainable = False
    style_outputs = [base_model.get_layer(name).output for name in style_layers]
    content_outputs = [base_model.get_layer(name).output for name in content_layers]
    model_outputs = style_outputs + content_outputs
    return models.Model(base_model.input, model_outputs)

# Define loss functions
def gram_matrix(input_tensor):
    channels = int(input_tensor.shape[-1])
    a = tf.reshape(input_tensor, [-1, channels])
    n = tf.shape(a)[0]
    gram = tf.matmul(a, a, transpose_a=True)
    return gram / tf.cast(n, tf.float32)

def style_loss(style, generated):
    return tf.reduce_mean(tf.square(gram_matrix(style) - gram_matrix(generated)))

def content_loss(content, generated):
    return tf.reduce_mean(tf.square(content - generated))

# Combine style and content losses
def total_loss(style_outputs, content_outputs, generated_outputs, style_weight=1e-2, content_weight=1e4):
    style_loss_value = 0
    content_loss_value = 0

    for style, generated in zip(style_outputs, generated_outputs[:len(style_outputs)]):
        style_loss_value += style_loss(style[0], generated[0])

    for content, generated in zip(content_outputs, generated_outputs[len(style_outputs):]):
        content_loss_value += content_loss(content[0], generated[0])

    style_loss_value *= style_weight / len(style_layers)
    content_loss_value *= content_weight / len(content_layers)

    return style_loss_value + content_loss_value

# Load and preprocess images
def load_and_process_img(img_path, img_size=(256, 256)):
    img = load_img(img_path, target_size=img_size)
    img = img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = tf.keras.applications.vgg19.preprocess_input(img)
    return img

def deprocess_img(processed_img):
    x = processed_img.copy()
    if len(x.shape) == 4:
        x = np.squeeze(x, 0)
    assert len(x.shape) == 3, ("Input to deprocess image must be an image of "
                             "dimension [1, height, width, channel] or [height, width, channel]")
    if len(x.shape) != 3:
        raise ValueError("Invalid input to deprocessing image")
  
    # Perform the inverse of the preprocessing step
    x[:, :, 0] += 103.939
    x[:, :, 1] += 116.779
    x[:, :, 2] += 123.68
    x = x[:, :, ::-1]

    x = np.clip(x, 0, 255).astype('uint8')
    return x

# Style Transfer function
def style_transfer(content_path, style_path, img_size=(256, 256), epochs=10, steps_per_epoch=100):
    content_img = load_and_process_img(content_path, img_size)
    style_img = load_and_process_img(style_path, img_size)
    generated_img = tf.Variable(content_img, dtype=tf.float32)

    model = get_model()
    style_outputs = model(style_img)
    content_outputs = model(content_img)
    generated_outputs = model(generated_img)

    optimizer = tf.optimizers.Adam(learning_rate=0.02, beta_1=0.99, epsilon=1e-1)

    for epoch in range(epochs):
        with tf.GradientTape() as tape:
            generated_outputs = model(generated_img)
            loss = total_loss(style_outputs, content_outputs, generated_outputs)

        grad = tape.gradient(loss, generated_img)
        optimizer.apply_gradients([(grad, generated_img)])

        # Clip the pixel values to the valid range [0, 255]
        generated_img.assign(tf.clip_by_value(generated_img, clip_value_min=0.0, clip_value_max=255.0))

        print(f"Epoch {epoch + 1}/{epochs}, Loss: {loss}")

    return generated_img.numpy()

# Example usage
content_image_path =  'C:\\Users\\Sathwika\\Downloads\\abstratc.png'

style_image_path = 'C:\\Users\\Sathwika\\Downloads\\pink.jpg'

stylized_image = style_transfer(content_image_path, style_image_path, epochs=10, steps_per_epoch=100)

# Display the images
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
ax1.imshow(load_and_process_img(content_image_path)[0])
ax1.set_title('Content Image')
ax1.axis('off')

ax2.imshow(load_and_process_img(style_image_path)[0])
ax2.set_title('Style Image')
ax2.axis('off')

ax3.imshow(deprocess_img(stylized_image))
ax3.set_title('Stylized Image')
ax3.axis('off')

plt.show()
```
